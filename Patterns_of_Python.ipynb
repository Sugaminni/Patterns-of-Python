Looping through a  list

```python
numbers = [1, 2, 3, 4]

# Takes each item and stores it in n. Then, prints out the numbers 1 at a time in each space
for n in numbers:
    print(n)
```

Accumulation: Used to compute totals, averages, gradients, etc.

```python
numbers = [1, 2, 3, 4]

total = 0

# Takes all the numbers and adds them (0 + 1 = 1 + 2 = 3 .... = 10)
for n in numbers:
    total += n

print(total)
```

Building a list: Creating a list from an existing one

```python
numbers = [1, 2, 3]

result = []

# Makes a new list of numbers using the existing numbers and multiplying by 2
for n in numbers:
    result.append(n * 2)

print(result)
```

Conditional Logic: Used when filtering data

```python
numbers = [1,2,3,4,5,6]

for n in numbers:
    if n % 2 == 0:
        print(n)
```

Filters, multiplies, and prints the squared list from existing one

```python
numbers = [1,2,3,4,5,6]

squaredlist = []

for n in numbers:
    if n % 2 == 0:
      squaredlist.append(n*n)

print(squaredlist)
```

Index-Based Loop: Takes the position in the list and prints out position : number

i = position


numbers[i] = value

```python
numbers = [10, 20, 30]

for i in range(len(numbers)):
    print(i, numbers[i])
```

Zip: Parallel Iteration

Basically manual vector multiplcation. In ML; becomes dot product

```python
x = [1,2,3]
w = [0.2,0.5,0.3]

# Multiply together
for xi, wi in zip(x, w):
    print (xi * wi)
```
